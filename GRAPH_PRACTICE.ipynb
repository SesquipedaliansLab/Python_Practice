```python
import csv as c
import matplotlib.pyplot as plt
#import numpy as np
#from scipy.interpolate import spline

f =open('temp.csv')
data = c.reader(f)
birth = input("Your BirthDay(XXXX-XX-XX): ")
birth = birth.split('-')
high = []
low = []
for row in data :
    date = row[0].split('-')
    if int(date[0])>=int(birth[0]) and date[1] == birth[1] and date[2] == birth[2] :
        high.append(float(row[-1]))
        low.append(float(row[-2]))
f.close()
% matplotlib inline
plt.rc('font', family = 'Consolas')
plt.title("Graph of Highest/Lowest Temperature at your birthday since your birth")
plt.plot(high, label = 'Highest temperature')
plt.plot(low, label = 'Lowest temperature')
xnew = np.linspace((high.min + low.min)/2,(high.max + low.max)/2,300)

tend = []
i = 0
for i in range(len(high)-1) :
    tend.append((high[i] + low[i])/2)

plt.plot(tend, label = "Tendency",markevery = "100")
plt.xticks(range(0,len(high),3), range(int(birth[0]), int(birth[0]) + len(high)+1, 3),rotation = 45)
plt.xlim(0, len(high))
plt.legend()
plt.show()
```
