```python
import csv
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

f = open("Thunder.csv")
data = csv.reader(f)
next(data)
data = list(data)
img = plt.imread("한반도지도_2.png")

Xaxis = []
Yaxis = []
Power = []
Time = []

iteration = 0
for row in data :
    Xaxis.append(float(row[2]))
    Yaxis.append(float(row[1]))
    Power.append(float(row[3]))
    Time.append(iteration)
    
for i in range(0,len(Power)) :
    Power[i] = abs(Power[i]/20)
    #Time[i] = '{0:06X}'.format(Time[i])

fig,ax = plt.subplots()
#plt.figure(dpi=300)
ax.imshow(img,extent = (123.8,134.8,31.3,43.3))

colors = cm.rainbow(np.linspace(0, 1, len(Time)))

#plt.xlim(121,134)
#plt.ylim(31,40.7)
plt.rc('font', family = "Malgun Gothic")
plt.title("한국의 낙뢰 분포도(1~6월)")
plt.xlabel('경도')
plt.ylabel("위도")

plt.scatter(Xaxis, Yaxis, s = Power, alpha = 0.2, c = colors)
plt.show()
```
