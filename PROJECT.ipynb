```python
import csv
import matplotlib.pyplot

f = open('FIWERT.csv')
AveSeatPerHour = [10800, 6750, 6171, 7363]

#자료
placeHolder = []
data = csv.reader(f)
next(data)
data = list(data)

#질문
date = input("날짜를 입력해주세요(XXXX-XX-XX): ")
time = input("타실 시간을 입력해주세요 : ")
station = input("타실 역을 입력해주세요 : ")
line = input("몇호선에서 타실 건가요? : ")

#빈칸을 모두 0으로
for Drow in data :
    for i in range(4,28) :
        if Drow[i] in (None,"") :
            Drow[i] = 0

# 질문한 것과 다르면 넘어감, 맞는 셀 2개 찾기
rowCountTmp = 0
rowCount = 0
for row in data :
    rowCountTmp += 1
    if date not in row[0] :
        continue
    if station not in row[2] :
        continue
    if line not in row[1] :
        continue
    placeHolder.append([row[2],row[4+int(time)],rowCountTmp])
    rowCount = rowCountTmp

#사람 수 세기
estimatedPeople = 0
startingJumper = rowCount % 40
repeatForColumn = rowCount // 40
Saved = 0
for Crow in data :
    for i in range(Saved, startingJumper) :
        if i % 2 is 0 :
            estimatedPeople += int(Crow[4+int(time)-repeatForColumn])
            print(estimatedPeople)
        else :
            estimatedPeople -= int(Crow[4+int(time)-repeatForColumn])
            print(estimatedPeople)
        Saved = i+1
    break
for Crow in data :
    for j in range(1, repeatForColumn) :
        for k in range(0, 40) :
            if k%2 is 0 :
                estimatedPeople += int(Crow[4+int(time)-repeatForColumn + j])
            else :
                estimatedPeople -= int(Crow[4+int(time)-repeatForColumn + j])
    break

#출력
print(rowCountTmp)
print(rowCount)
print(startingJumper)
print(repeatForColumn)
print(placeHolder)
print(estimatedPeople)
```
