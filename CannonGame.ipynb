```python
import turtle as t
import random
t.title("CannonGame")
target = 0
def ChooseTarget() :
    target = random.randint(50,150)


#draw stage
def start() :
    t.pensize(1)
    t.showturtle()
    t.clear()
    t.goto(-300,0)
    t.down()
    t.goto(300,0)
    t.pensize(3)
    t.up()
    t.goto(target - 25, 2)
    t.down()
    t.goto(target + 25, 2)
    t.color("black")
    t.up()
    t.goto(0,-100)
    t.write("Press up and down to aim, space to fire",False,"center",(" ",15))
    t.goto(-200,10)
    t.setheading(20)
    t.pensize(1)

#turn up the cursor
def turn_up() :
    if(t.heading()!=90) :
        t.left(2)
#turn down the cursor
def turn_down() :
    if(t.heading()!=0) :
        t.right(2)
#fire
def fire() :
    angle = t.heading()
    while t.ycor()>0 :
        t.forward(15)
        t.right(5)
    d = t.distance(target,0)
    t.sety(random.randint(10,100))
    if d<25 :
        t.color("blue")
        t.write("Good", False, "center", (" ", 15))
        t.hideturtle()
        t.goto(0,-200)
        t.write("Press [Enter] to restart", False,"center",(" ", 15))
        t.color("black")
        t.pensize(1)
    else :
        t.color("red")
        t.write("Miss",False,"center", (" ", 15))
        t.color("black")
        t.goto(-200,10)
        t.setheading(angle)
        
start()

t.onkeypress(turn_up,"Up")
t.onkeypress(turn_down,"Down")
t.onkeypress(fire, "space")
t.onkeypress(start,"Return")
t.listen()
t.mainloop()
```
